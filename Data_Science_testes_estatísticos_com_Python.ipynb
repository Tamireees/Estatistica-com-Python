```python
import numpy as np
# importaremos o numpy (import numpy as np) e passaremos uma seed aleatória ("semente", em inglês)
# que nos permita fixar uma distribuição (de modo que o processo seja reproduzível).
```

```python
# No Python, existe uma maneira mais direta de criarmos uma lista na qual os elementos são adicionados um a um
# progressivamente: basta passarmos as condições a serem executadas no for antes desse iterador, e colocarmos
# toda essa construção entre colchetes:
```

```python
medias = [temp[0:i].mean() for i in range (1, len(nota_media_dos_filmes_com_pelo_menos_10_voto))]

plt.plot(medias)
```

```python
# O ztest() nos devolve duas variáveis: o valor estatístico e o pvalue. Repare que este último é um valor bastante baixo, e bem menor que o nosso 0.05
# (o alpha). Portanto, podemos descartar a hipótese de que, no mundo real, a média do Toy Story é igual a 3.4320503405352603 (também chamada de hipótese nula).
```

```python
np.random.seed(75243)
temp = notas1.sample(frac=1).rating
temp
```

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

medias = [temp[0:i].mean() for i in range(1, len(temp))]

plt.plot(medias)
```

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

def calcula_teste(i):
  media = temp[0:i].mean()
  stat, p = ztest(temp[0:i], value= 3.4320503405352594)
  return(i, media, p)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])

plt.plot(valores[:,0], valores[:,1])
plt.plot(valores[:,0], valores[:,2])
plt.hlines(y= 0.05, xmin= 2, xmax= len(temp), colors='r')
```

```python
# Nessa situação, poderia-se afirmar, a partir do Teste Z:
#que não há evidências para descartar a hipótese nula
#que não há dados suficientes para tirar essa conclusão
```

```python
plt.boxplot([notas1.rating, notas593.rating, notas72226.rating], labels=["Toy Story", "Silence of the Lambs", "Fantastic Mr. Fox"])

plt.title("Distribuição das notas de acordo com os filmes")
```

```python
notas72226.describe()
```

```python
sns.boxplot(x = "movieId", y = "rating", data = notas.query("movieId in (1, 593, 72226)"))
```

```python
#Dessa forma, o sns.boxplot() é interessante quando nossos dados podem ser facilmente separados por colunas. No caso do plt.boxplot(),
#temos que separar esses dados manualmente.
```

```python
descr_1 = DescrStatsW(notas1.rating)
descr_593 = DescrStatsW(notas593.rating)

comparacao = descr_1.get_compare(descr_593)

comparacao.summary()
```

```python
# "Aplicando o Teste Z com alpha = 0.05, encontramos uma diferença estatística entre os filmes Toy Story e Silence of the Lambs,
# com um p value igual a 0.02 e um intervalo de confiança entre -0.391 e -0.090".
```

```python
descr_72226 = DescrStatsW(notas72226.rating)
descr_593 = DescrStatsW(notas593.rating)

comparacao = descr_72226.get_compare(descr_593)

comparacao.summary(use_t=True)
```

```python
# Dessa vez, nosso p value não indica uma diferença significante (pois não é menor do que 0.05), e o intervalo inclui tanto a chance do filme ser pior,
# quanto a chance dele ser melhor.
```

```python
comparacao = descr_1.get_compare(descr_72226)

comparacao.summary(use_t=True)
```

```python
# Da mesma forma, não temos uma diferença significativa, e o intervalo também inclui aquelas duas possibilidades.
# Portanto, a única diferença que encontramos nos nossos dados foi entre os filmes Toy Story e Silence of the Lambs - na qual o último
# filme é melhor (talvez um pouco melhor, talvez razoavelmente melhor).
```

# Comparação não paramétrica com Wilcoxon Ranksums

```python
from scipy.stats import normaltest
```

```python
stat, p = normaltest(notas1.rating)
p
# o P_value é menor que 0.05, ou seja, nao é uma distribuiçao normal
```

```python
# Teremos um p value de 0.00011053430732728716 - ou seja, menor do que 0.05. Como essa função testa a hipótese nula de que o conjunto de dados
# venha de uma distribuição normal, um resultado menor que 0.05 indica que não estamos trabalhando com uma distribuição normal.

# Isso também significa que não poderíamos aplicar o Teste T ou o Teste Z, mas sim testes que funcionam com outros tipos de distribuição.
# Nos cursos de estatística da Alura, abordamos testes que funcionam nessas situações: os testes não paramétricos.
```

```python
from scipy.stats import ranksums

_, p = ranksums(notas1.rating, notas593.rating)
p
```
