```python
import numpy as np
# importaremos o numpy (import numpy as np) e passaremos uma seed aleatÃ³ria ("semente", em inglÃªs)
# que nos permita fixar uma distribuiÃ§Ã£o (de modo que o processo seja reproduzÃ­vel).
```

```python
np.random.seed(75243)

temp = nota_media_dos_filmes_com_pelo_menos_10_voto.sample(frac=1)

medias = list()
for i in range (1, len(nota_media_dos_filmes_com_pelo_menos_10_voto)):
  medias.append(temp[0:i].mean())

plt.plot(medias)
```

```python
# No Python, existe uma maneira mais direta de criarmos uma lista na qual os elementos sÃ£o adicionados um a um
# progressivamente: basta passarmos as condiÃ§Ãµes a serem executadas no for antes desse iterador, e colocarmos
# toda essa construÃ§Ã£o entre colchetes:
```

```python
medias = [temp[0:i].mean() for i in range (1, len(nota_media_dos_filmes_com_pelo_menos_10_voto))]

plt.plot(medias)
```

```python
# O ztest() nos devolve duas variÃ¡veis: o valor estatÃ­stico e o pvalue. Repare que este Ãºltimo Ã© um valor bastante baixo, e bem menor que o nosso 0.05
# (o alpha). Portanto, podemos descartar a hipÃ³tese de que, no mundo real, a mÃ©dia do Toy Story Ã© igual a 3.4320503405352603 (tambÃ©m chamada de hipÃ³tese nula).
```

```python
np.random.seed(75243)
temp = notas1.sample(frac=1).rating
temp
```

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

medias = [temp[0:i].mean() for i in range(1, len(temp))]

plt.plot(medias)
```

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

def calcula_teste(i):
  media = temp[0:i].mean()
  stat, p = ztest(temp[0:i], value= 3.4320503405352594)
  return(i, media, p)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])

plt.plot(valores[:,0], valores[:,1])
plt.plot(valores[:,0], valores[:,2])
plt.hlines(y= 0.05, xmin= 2, xmax= len(temp), colors='r')
```

```python
# Nessa situaÃ§Ã£o, poderia-se afirmar, a partir do Teste Z:
#que nÃ£o hÃ¡ evidÃªncias para descartar a hipÃ³tese nula
#que nÃ£o hÃ¡ dados suficientes para tirar essa conclusÃ£o
```

# ComparaÃ§Ã£o nÃ£o paramÃ©trica com Wilcoxon Ranksums

```python
from scipy.stats import normaltest
```

```python
stat, p = normaltest(notas1.rating)
p
# o P_value Ã© menor que 0.05, ou seja, nao Ã© uma distribuiÃ§ao normal
```

```python
# Teremos um p value de 0.00011053430732728716 - ou seja, menor do que 0.05. Como essa funÃ§Ã£o testa a hipÃ³tese nula de que o conjunto de dados
# venha de uma distribuiÃ§Ã£o normal, um resultado menor que 0.05 indica que nÃ£o estamos trabalhando com uma distribuiÃ§Ã£o normal.

# Isso tambÃ©m significa que nÃ£o poderÃ­amos aplicar o Teste T ou o Teste Z, mas sim testes que funcionam com outros tipos de distribuiÃ§Ã£o.
# Nos cursos de estatÃ­stica da Alura, abordamos testes que funcionam nessas situaÃ§Ãµes: os testes nÃ£o paramÃ©tricos.
```
